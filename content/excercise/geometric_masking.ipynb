```python
# Creating a Geometric Mask with Broadcasting
# Your task is to create a 2D "target" mask, where a circular ring (annulus) is marked as 1 and everything else is 0. This is useful for selecting regions of interest in images.

# Your Task:

# Create a 1D array for x and y coordinates from -5 to 5 with 50 points using np.linspace.
# Use np.meshgrid to create 2D coordinate grids X and Y.
# Calculate a 2D array R representing the distance of each point (X, Y) from the origin (0, 0). The formula is 
# . This calculation will rely on broadcasting.
# Create a new array target_mask of the same shape as R, initialized with zeros.
# Use masked indexing with logical operators to set the elements of target_mask to 1 where R is between an inner radius of 2.0 and an outer radius of 4.0 (i.e., 2.0 < R < 4.0).
# Use matplotlib.pyplot.imshow() to visualize your final target_mask.



import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create coordinate arrays
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

# Step 2: Create 2D grid with meshgrid
X, Y = np.meshgrid(x, y)

# Step 3: Compute distance from origin (broadcasting)
R = np.sqrt(X**2 + Y**2)

# Step 4: Initialize target mask
target_mask = np.zeros_like(R)

# Step 5: Apply logical mask for annulus (2.0 < R < 4.0)
mask = (R > 2.0) & (R < 4.0)
target_mask[mask] = 1

# Step 6: Visualize
plt.imshow(target_mask, extent=(-5, 5, -5, 5), origin="lower", cmap="gray")
plt.colorbar(label="Mask Value")
plt.title("Circular Ring Mask (Annulus)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()



# np.meshgrid → creates X, Y coordinate grids.

# R = sqrt(X² + Y²) → computes radial distance for each grid point.

# Boolean mask (R > 2.0) & (R < 4.0) → selects annulus points.

# Assigning target_mask[mask] = 1 → marks the circular ring region.

# imshow displays the mask (bright = 1, dark = 0).
```
