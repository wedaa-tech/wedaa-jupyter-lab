```python
import numpy as np
import matplotlib.pyplot as plt


# # Creating three arrays similar to the example
# runs_a = np.array([1, 19, 58, 2, 11, 4, 2, 0, 12, 33, 27, 0, 100, 4, 5])
# runs_b = np.array([15, 3, 45, 7, 22, 0, 8, 14, 29, 6, 41, 2, 88, 10, 1])
# runs_c = np.array([30, 5, 63, 9, 17, 12, 3, 25, 8, 44, 16, 1, 92, 6, 20])

# # Calculating means
# means = [np.mean(runs_a), np.mean(runs_b), np.mean(runs_c)]

# # Creating bar chart
# plt.figure(figsize=(8, 6))
# plt.bar(['Array A', 'Array B', 'Array C'], means, color=['blue', 'green', 'red'])
# plt.title('Mean Values of Three Arrays')
# plt.ylabel('Mean Value')
# plt.xlabel('Arrays')
# plt.grid(axis='y', linestyle='--', alpha=0.7)
# plt.show()


dhoni = np.loadtxt("data/dhoni.txt")
sachin = np.loadtxt("data/sachin.txt")
kohli = np.loadtxt("data/kohli.txt")
kapil = np.loadtxt("data/kapil_dev.txt")
gambhir = np.loadtxt("data/gambhir.txt")
rohit = np.loadtxt("data/rohit.txt")
sehwag = np.loadtxt("data/sehwag.txt")


dhoni_mean = np.mean(dhoni)
print(dhoni_mean)
dhoni_median = np.median(dhoni)
print(dhoni_median)
print(max(dhoni))
dhoni_mod = np.mod(dhoni, 100)
#print(dhoni_mod)



means =[np.mean(dhoni), np.mean(sachin), np.mean(kohli)]
std_dev = [np.std(dhoni, ddof=1), np.std(sachin, ddof=1), np.std(kohli, ddof=1)]

plt.bar(['dhoni', 'sachin', 'kohli'], means, yerr=std_dev, color=['red','green','blue'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title('Mean Values of Three players')
plt.ylabel('Mean Value')
plt.xlabel('Arrays')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# Creating two subplots for histogram comparisons
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))


# Plot 1: Comparing dhoni and sachin
ax1.hist(dhoni, bins=10, alpha=0.5, label='Dhoni', color='blue')
ax1.hist(sachin, bins=10, alpha=0.5, label='Sachin', color='green')
ax1.set_title('Histogram: Dhoni vs Sachin')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')
ax1.legend()
ax1.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.boxplot([dhoni], vert=True, patch_artist=True, showfliers=True)
plt.title("Boxplot for player runs")
plt.ylabel("Runs")
plt.xticks([1], ["Match scores"])
plt.show()


plt.figure(figsize=(10,6))
scores = [dhoni, sachin, kohli]

plt.boxplot(scores, labels=['dhoni', 'sachin', 'kohli'], patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='blue'), medianprops=dict(color='red'),
           whiskerprops=dict(color='blue'),
           capprops=dict(color='blue'))

plt.title('Boxplot Comparison of Dohni, Sachin, Kohli')
plt.ylabel('Runs')
plt.xlabel('Players')
plt.grid(True, alpha=0.7)
plt.show()

plt.violinplot(scores)
plt.show()


plt.figure(figsize=(10, 6))
plt.hist(dhoni, bins=10, alpha=0.4, label='Dhoni', color='blue')
plt.hist(sachin, bins=10, alpha=0.4, label='Sachin', color='green')
plt.hist(kohli, bins=10, alpha=0.4, label='Kohli', color='red')
plt.hist(sehwag, bins=10, alpha=0.4, label='Sehwag', color='purple')
plt.title('Histogram Comparison of All Four players')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# plt.boxplot(data, labels=['Array A', 'Array B', 'Array C', 'Array D'], patch_artist=True, 
#             boxprops=dict(facecolor='lightblue', color='blue'),
#             medianprops=dict(color='red'),
#             whiskerprops=dict(color='blue'),
#             capprops=dict(color='blue'))
# plt.title('Boxplot Comparison of Four Arrays')
# plt.ylabel('Value')
# plt.xlabel('Arrays')
# plt.grid(True, axis='y', linestyle='--', alpha=0.7)
# plt.show()



```
