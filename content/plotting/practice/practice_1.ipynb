Problem 1:
Simulate 1000 tosses of a biased coin that shows heads (H) with probability 0.7 and tails (T) with probability 0.3.

Plot a histogram of the outcomes.

Compare the observed probabilities with the theoretical probabilities (0.7 and 0.3).


```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_tosses = 1000
p_heads = 0.7  # Probability of heads
p_tails = 0.3  # Probability of tails

# Simulate coin tosses
tosses = np.random.choice(['H', 'T'], size=n_tosses, p=[p_heads, p_tails])

# Count occurrences
heads_count = np.sum(tosses == 'H')
tails_count = np.sum(tosses == 'T')

# Calculate observed probabilities
obs_p_heads = heads_count / n_tosses
obs_p_tails = tails_count / n_tosses

# Print comparison of observed vs theoretical probabilities
print(f"Number of Heads: {heads_count}")
print(f"Number of Tails: {tails_count}")
print(f"Observed Probability of Heads: {obs_p_heads:.3f} (Theoretical: {p_heads})")
print(f"Observed Probability of Tails: {obs_p_tails:.3f} (Theoretical: {p_tails})")
print(f"Difference in Heads Probability: {abs(obs_p_heads - p_heads):.3f}")
print(f"Difference in Tails Probability: {abs(obs_p_tails - p_tails):.3f}")

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Heads', 'Tails'], [heads_count, tails_count], color=['blue', 'green'])
plt.title('Results of 1000 Biased Coin Tosses (P(H) = 0.7, P(T) = 0.3)')
plt.ylabel('Frequency')
plt.xlabel('Outcome')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Create subplots for heads and tails
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Heads histogram
ax1.hist(tosses[tosses == 'H'], bins=1, rwidth=0.4, color='blue', alpha=0.7)
ax1.set_title('Histogram of Heads')
ax1.set_xlabel('Outcome')
ax1.set_ylabel('Frequency')
ax1.set_xticks(['H'])
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Subplot 2: Tails histogram
ax2.hist(tosses[tosses == 'T'], bins=1, rwidth=0.4, color='green', alpha=0.7)
ax2.set_title('Histogram of Tails')
ax2.set_xlabel('Outcome')
ax2.set_ylabel('Frequency')
ax2.set_xticks(['T'])
ax2.grid(axis='y', linestyle='--', alpha=0.7)

plt.suptitle('Histograms of 1000 Biased Coin Tosses (P(H) = 0.7, P(T) = 0.3)', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()


# Create histogram for both heads and tails
plt.figure(figsize=(8, 6))
plt.hist([tosses[tosses == 'H'], tosses[tosses == 'T']], bins=2, label=['Heads', 'Tails'], 
         color=['blue', 'green'], alpha=0.7, rwidth=0.45)
plt.title('Histogram of 1000 Biased Coin Tosses (P(H) = 0.7, P(T) = 0.3)')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.xticks(['H', 'T'])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```
