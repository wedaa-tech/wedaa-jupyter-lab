Data Set:  65, 72, 78, 80, 82, 85, 88, 90, 92, 95, 100, 102, 105, 120, 150

#### Step-by-Step Manual Calculation:

Sort the Data:
[65, 72, 78, 80, 82, 85, 88, 90, 92, 95, 100, 102, 105, 120, 150]
Calculate the Median (Q2):

$ n = 15 $ (odd), so the median is the value at position $ \frac{n + 1}{2} = \frac{15 + 1}{2} = 8 $.
Median = 90 (8th value).


#### Calculate Quartiles (Q1 and Q3):

Q1 (25th percentile): Split the data into the lower half (first 7 values: [65, 72, 78, 80, 82, 85, 88]).

Median of lower half (7 values, so position $ \frac{7 + 1}{2} = 4 $) = 80.


Q3 (75th percentile): Upper half (last 7 values: [92, 95, 100, 102, 105, 120, 150]).

Median of upper half (position 4) = 102.


So, Q1 = 80, Q3 = 102.


#### Calculate IQR:
$$\text{IQR} = Q3 - Q1 = 102 - 80 = 22$$

#### Determine Whisker Boundaries:

Lower bound: $ Q1 - 1.5 \times \text{IQR} = 80 - 1.5 \times 22 = 80 - 33 = 47 $.
Upper bound: $ Q3 + 1.5 \times \text{IQR} = 102 + 1.5 \times 22 = 102 + 33 = 135 $.
Whiskers extend to the smallest and largest values within these bounds:

Lower whisker: Smallest value ≥ 47 = 65.
Upper whisker: Largest value ≤ 135 = 120.




#### Identify Potential Outliers:

Values below 47 or above 135 are outliers.
Check the dataset: 150 > 135, so 150 is a potential outlier.



#### Box Plot Description:

Box: Extends from Q1 (80) to Q3 (102), with a line at the median (90).
Whiskers: Extend from 65 (smallest value ≥ 47) to 120 (largest value ≤ 135).
Outlier: 150 is plotted as a separate point.





```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
data = [65, 72, 78, 80, 82, 85, 88, 90, 92, 95, 100, 102, 105, 120, 150]

# Create box plot
plt.boxplot(data, vert=True, patch_artist=True, showfliers=True)
plt.title("Box Plot of Exam Scores")
plt.ylabel("Scores")
plt.xticks([1], ["Exam Scores"])

# Add annotations for quartiles, median, and whiskers
median = np.median(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
lower_whisker = min([x for x in data if x >= q1 - 1.5 * iqr])
upper_whisker = max([x for x in data if x <= q3 + 1.5 * iqr])

plt.text(1.1, median, f'Median: {median}', verticalalignment='center', color='red')
plt.text(1.1, q1, f'Q1: {q1}', verticalalignment='center', color='blue')
plt.text(1.1, q3, f'Q3: {q3}', verticalalignment='center', color='blue')
plt.text(1.1, lower_whisker, f'Lower Whisker: {lower_whisker}', verticalalignment='center', color='green')
plt.text(1.1, upper_whisker, f'Upper Whisker: {upper_whisker}', verticalalignment='center', color='green')
plt.text(1.1, 150, f'Outlier: 150', verticalalignment='center', color='purple')

# Show plot
plt.show()
```
