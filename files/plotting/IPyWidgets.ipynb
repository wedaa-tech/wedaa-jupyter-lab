```python
import numpy as np
import matplotlib.pyplot as plt

def sine_plot(k=2, phi=0.5):
    # Create x values from 0 to 2π
    x = np.linspace(0, 2*np.pi, 1000)
    
    # Calculate sine with given parameters
    y = np.sin(k*x + phi)
    
    # Create the plot
    plt.figure(figsize=(8, 4))
    plt.plot(x, y)
    plt.title(f'Sine Wave: sin({k}x + {phi})')
    plt.xlabel('x')
    plt.ylabel('sin(kx + φ)')
    plt.grid(True)
    plt.show()
    
    # Return values if needed
    return x, y

# Call the function
sine_plot(k=2, phi=0.5)
```
