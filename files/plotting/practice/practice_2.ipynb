Problem 2: Simulate rolling a fair die 6000 times.

Plot the histogram of outcomes.

Compute observed probabilities of each face (1–6).

Compare with the theoretical probability (1/6 ≈ 0.1667).

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_rolls = 6000
faces = [1, 2, 3, 4, 5, 6]
p_fair = [1/6] * 6  # Equal probability for each face

# Simulate die rolls
rolls = np.random.choice(faces, size=n_rolls, p=p_fair)

# Count occurrences of each face
counts = [np.sum(rolls == face) for face in faces]
#counts = [rolls.count(face) for face in faces]

#counts = []
#for face in faces:
#    counts.append(rolls.count(face))
    

# Print counts and observed probabilities
for face, count in zip(faces, counts):
    print(f"Face {face}: {count} times (Observed Probability: {count/n_rolls:.3f}, Theoretical: {1/6:.3f})")

# Print counts and observed probabilities
# for i in range(len(faces)):
#     face = faces[i]
#     count = counts[i]
#     print(f"Face {face}: {count} (Probability: {count/n_rolls:.3f})")
    
# Create histogram
plt.figure(figsize=(8, 6))
plt.hist(rolls, bins=range(1, 8), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.title('Histogram of 6000 Fair Die Rolls')
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.xticks(faces)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
