```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data/data.txt")
print(len(data))
data[:2]
```

```python
x,y = np.loadtxt('data/data.txt', unpack=True)

print(len(x), len(y))
print(x[1])
mean_x = np.mean(x)
print("Mean X:", mean_x)
mean_y = np.mean(y)
print("Mean Y:", mean_y)

## Baised_Estimator
standard_deviation_x = np.std(x)
standard_deviation_y = np.std(y)
print(standard_deviation_x)
print(standard_deviation_y)

## UnBaised estimator
u_standard_deviation_x = np.std(x, ddof=1)
print(u_standard_deviation_x)
u_standard_deviation_y = np.std(y, ddof=1)
print(u_standard_deviation_y)
```

```python
percentile_x = np.percentile(x, 50)
print(percentile_x)

percentile_in_oneShot = np.percentile(x,[25,50,75,90])
print(percentile_in_oneShot)
```

```python
percentiles = np.arange(0,101,1)
percentile_values = np.percentile(x, percentiles)
print(percentile_values)
```

```python
plt.plot(percentiles, percentile_values, marker="o", markersize=2, linestyle="-")
plt.title("Percentile of X [0-100]")
plt.xlabel("Percentile")
plt.ylabel("Values at percentile")
plt.grid(True)
plt.show()
```
