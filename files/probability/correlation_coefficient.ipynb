```python
import numpy as np

# Data
x = np.array([1, 2, 3])
y = np.array([2, 4, 6])

# Calculate means
x_mean = np.mean(x)
y_mean = np.mean(y)

# Compute deviations and products
x_dev = x - x_mean
y_dev = y - y_mean
cov_term = np.sum(x_dev * y_dev)
x_var = np.sum(x_dev**2)
y_var = np.sum(y_dev**2)

# Pearson correlation coefficient
r = cov_term / np.sqrt(x_var * y_var)

print("Sample correlation coefficient (mean-based formula):", r)
```

```python
# Using corrcoef
r_numpy = np.corrcoef(x, y)[0, 1]
print("Sample correlation coefficient (NumPy corrcoef):", r_numpy)
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]

plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y, color='red', linestyle='--', label='Line y=2x')
plt.title('Scatter Plot of Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Data
x = [1, 2, 3]
y = [2, 4, 6]

# Compute correlation matrix
corr_matrix = np.corrcoef(x, y)
print("Correlation matrix:\n", corr_matrix)

# Extract correlation coefficient between x and y
r_numpy = corr_matrix[0, 1]
print("Correlation coefficient (x, y):", r_numpy)
```
