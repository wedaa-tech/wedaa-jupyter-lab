```python
# Problem Statement

# Consider the 7 different batsman data provided to you.

# Compute the mean, median, mode, standard deviation for each of these batsmen.

# Store the names of the batsmen in a list.

# Then store the mean/median/mode for each batsman in the same order.

# You should have 5 lists with you at the end of this.


import numpy as np
from scipy import stats

# Example batsman data (replace with actual data you have)
batsman_data = {
    "Batsman1": [45, 67, 23, 89, 55, 67, 42],
    "Batsman2": [10, 15, 20, 10, 25, 30, 10],
    "Batsman3": [50, 50, 50, 60, 70, 80, 90],
    "Batsman4": [100, 95, 110, 120, 115, 105, 100],
    "Batsman5": [5, 15, 25, 35, 45, 55, 65],
    "Batsman6": [0, 0, 5, 10, 0, 15, 20],
    "Batsman7": [33, 44, 55, 66, 77, 88, 99],
}

# Lists to store results
names = []
means = []
medians = []
modes = []
std_devs = []

# Compute statistics for each batsman
for name, scores in batsman_data.items():
    scores_array = np.array(scores)

    names.append(name)
    means.append(np.mean(scores_array))
    medians.append(np.median(scores_array))

    # mode() in scipy returns ModeResult object → take mode[0]
    mode_val = stats.mode(scores_array, keepdims=True).mode[0]
    modes.append(mode_val)

    std_devs.append(np.std(scores_array, ddof=0))  # population std dev

# Final 5 lists
print("Names:", names)
print("Means:", means)
print("Medians:", medians)
print("Modes:", modes)
print("Standard Deviations:", std_devs)

```

Histogram of All 7 Batsmen Together

A histogram is best suited for showing the distribution of a single dataset.

If you try to plot all 7 batsmen together in one histogram:

You’d either merge their scores into one big distribution → this hides individual batsman differences.

Or you’d use overlapping histograms (different colors) → this quickly becomes messy and unreadable with 7 players.

Alternatively, you could use a stacked histogram, but interpretation is still hard when categories overlap.

Conclusion: Not a good idea, because histograms are not well-suited for direct comparison across many groups.

✅ Boxplot

Boxplots summarize median, quartiles, spread, and outliers for each batsman.

They work well when comparing multiple groups side-by-side.

For 7 batsmen, you get a clear, compact view of who has higher/lower medians, who is more consistent (narrow IQR), and who has extreme scores.

You lose detail about distribution shape (e.g., skewness, bimodality).

✅ Violin Plot

Violin plots combine boxplot summary with distribution shape (via KDE).

For 7 batsmen, this gives the best compromise: you see median + spread and whether their scores are symmetric, skewed, or have multiple peaks.

Slight downside: violin plots can be harder to interpret if the audience isn’t familiar with them.

✅ Final Contrast

Histogram (all together): Not suitable — cluttered, loses individual insights.

Boxplot: Great for quick statistical comparison across batsmen.

Violin plot: Even better if you want to show both summary stats and distribution shape for each batsman.

```python
import numpy as np
import matplotlib.pyplot as plt

# Example batsman data
batsman_data = {
    "Batsman1": [45, 67, 23, 89, 55, 67, 42],
    "Batsman2": [10, 15, 20, 10, 25, 30, 10],
    "Batsman3": [50, 50, 50, 60, 70, 80, 90],
    "Batsman4": [100, 95, 110, 120, 115, 105, 100],
    "Batsman5": [5, 15, 25, 35, 45, 55, 65],
    "Batsman6": [0, 0, 5, 10, 0, 15, 20],
    "Batsman7": [33, 44, 55, 66, 77, 88, 99],
}

names = list(batsman_data.keys())
scores = list(batsman_data.values())

plt.figure(figsize=(12, 6))

# ❌  Attempt 1: Overlayed histogram (cluttered)
plt.subplot(1, 2, 1)
for name, vals in batsman_data.items():
    plt.hist(vals, bins=10, alpha=0.5, label=name)
plt.title("Overlayed Histogram (Messy)")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.legend(fontsize=8)

# ✅ Attempt 2: Boxplot (clean)
plt.subplot(1, 2, 2)
plt.boxplot(scores, labels=names, patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="black"),
            medianprops=dict(color="red", linewidth=2))
plt.title("Boxplot Comparison (Clear)")
plt.ylabel("Runs")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()




# What You’ll See

# Left (Overlayed Histogram):

# Bars from different batsmen overlap.

# Colors mix → hard to interpret.

# Impossible to compare central tendency & spread clearly.

# Right (Boxplot):

# Each batsman’s median, spread, and outliers are shown neatly.

# Very easy to compare across 7 batsmen side by side.
```
